```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
taxi = pd.read_csv('nyc_taxis.csv')
```

```python
taxi.head()
```

```python
n_taxi = taxi.drop(['pickup_time','fare_amount','fees_amount','tolls_amount','tip_amount'],axis=1) #removed extra column
```

```python
n_taxi
n_taxi['trip_length'] = n_taxi['trip_length']/60 #converted trip_length(in seconds) into trip_lenth(in mins) 
```

```python
n_taxi.info() #to check the type of data
```

map() tool is used to update the DataFrame in column. First assigned values to the variable and then used map() tool to update the DataFrame.

```python
dmonth = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sept',10:'Oct',11:'Nov',12:'Dec'} 
#values assigned
```

```python
n_taxi['pickup_month'] = n_taxi['pickup_month'].map(dmonth) 
#updated the values of pickup_month column with dmonth values
```

```python
d_week = {1:'Mon',2:'Tue',3:'Wed',4:'Thu',5:'Fri',6:'Sat',7:'Sun'} #values assigned
```

```python
n_taxi['pickup_dayofweek'] = n_taxi['pickup_dayofweek'].map(d_week)
 #updated the values of pickup_dayofweek column with d_week values
```

```python
n_pay = {1:'Card',2:'Cash'} #values assigned to n_pay 
```

```python
n_taxi['payment_type'] = n_taxi['payment_type'].map(n_pay)
#updated the values of payment_type column with n_pay values.
```

```python
n_taxi
# updated DataFrame looks like this
```

```python
#renamed the column trip_length into time in mins
n_taxi.rename(columns={'trip_length':'time in mins'})
```

```python
import seaborn as sns
sns.set_style('darkgrid')
```

```python
sns.barplot(x= 'pickup_location_code',y='dropoff_location_code',hue='payment_type',data = n_taxi)
```

Using Seaborn, vizualised the pickup_location_code vs dropoff_location_code bar graph and made conclusion
that Most of the cabs pick their passengers from 'LOCATION_CODE=3' and drop them at 'LOATION_CODE=4'.

```python
sns.set_style('whitegrid')
sns.lineplot(x = 'pickup_month',y = 'pickup_location_code',data=n_taxi)
```

The above Line Plot brings out the conclusion that the during the month of Jan-Feb, the no. of passengers increases and then 
shows a zig-zag pattern
